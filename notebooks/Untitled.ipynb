```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import requests
from bs4 import BeautifulSoup as BS
import time
```

```python
URL = 'https://realpython.github.io/fake-jobs/'

response = requests.get(URL)
```

```python
response.status_code
```

```python
soup = BS(response.text)
```

```python
soup
```

```python
soup.find('h2').text
```

```python
job_titles = soup.findAll('h2')
type(job_titles)
```

```python
job_titles
```

```python
job_titles = [title.text.strip() for title in job_titles]
job_titles
```

```python
company = [c.text.strip() for c in soup.findAll('h3')]
company
```

```python
location = [l.text.strip() for l in soup.findAll('p', class_="location")]
location
```

```python
time = [t.text.strip() for t in soup.findAll('time')]
time
```

```python
jobs = pd.DataFrame({'job_title': job_titles, 'company': company, 'location': location, 'date': time})
jobs.head()
```

```python
urls = [u.get('href') for u in soup.findAll('a', string="Apply")]
urls
```
